### Question 1:
Write a definition for a function which takes two lists as input parameter and returns a single zipped list such that every i<sup>th</sup> item of both list will be enclosed in a tuple, so on and so forht to form a list of tuples to return.

for example, _x_ and _y_ are the lists, and _f_ is the definiton then the return value should be _z_
```
x = [1,2,3,4]
y = [5,6,7,8]
z = f(x,y)
print(z
[(1,5), (2,6), (3,7), (4,8)]
```

````python
def Zipped_List(list1,list2):
    zipped = []
    for index1 in range(len(list1)):
        for index2 in range(len(list2)):
            if (index1 == index2):
                zipped.append((list1[index1],list2[index2]))
    return zipped

def Zipped_List_With_Comprehension(l1,l2):
    z=[(l1[idx1],l2[idx2]) for idx1 in range(len(l1)) for idx2 in range(len(l2)) if idx1==idx2]
    return z;
    

x = [1,2,3,4,9,11]
y = [5,6,7,8,3,15]
z = Zipped_List(x,y)
print("Output without comprehension: ", z)

zc = Zipped_List_With_Comprehension(x,y)
print("Output with comprehension: ", zc)
````

- Implement a definition for the function _force_, i.e., force = mass * accelartion, this function will take both lists, i.e., mass and acceleration, and returns the list of forces w.r.t values of mass and acceleration of coresponding indicies. 

for example, _m_ and _a_ are the lists for mass and acceleration, and _force_ is the definiton for force then, _f_ is the resulting list of forces.
```
m = [1.3, 2.5, 3.4, 4.8]
a = [5.4, 6.2, 7.6, 8.1]
f = force(x,y)
print(f)
[7.0200000000000005, 15.5, 25.84, 38.879999999999995]
```

````python
import numpy as np

def Calculate_Force(m,a):
    f=[m[i]*a[j] for i in range(len(m)) for j in range(len(a)) if i==j]
    return f

#Example Given
m = [1.3, 2.5, 3.4, 4.8]
a = [5.4, 6.2, 7.6, 8.1]
f = Calculate_Force(m,a)
print("Given Example Output:", f)

#Generating random mass and acceleration
mlist = [round(np.random.uniform(1,100),1) for x in range(5)]
alist = [round(np.random.uniform(1,100),1) for y in range(5)]

print("\nRandom mass generation list:\t", mlist)
print("Random acceleration list:\t", alist)

flist = Calculate_Force(mlist,alist)
print("Output for force list:\t", flist)
````
