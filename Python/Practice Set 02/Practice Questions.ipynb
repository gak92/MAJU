####   1-Your task is to write this file into a json file format. The json file will look like:
####   { tweet_id : { sentiment: { Author : content}}}

```python
##########################################################
#         Solution to task 1
##########################################################
import json, csv

data = {}

with open('text_emotion.csv', 'r') as csvfile:
    csvreader = csv.DictReader(csvfile)
    for rows in csvreader:
        #print(rows)
        tweet_id = rows['tweet_id']
        sentiment = rows['sentiment']
        author = rows['author']
        content = rows['content']
        data[tweet_id] = {sentiment: {author: content}}
        #print(data)        
        
        
with open('output.json', 'w') as jsonfile:
    jsonfile.write(json.dumps(data, indent=4))
    
```

#### 2- After this, you have to make few functions to_Sentiment(json_file), to_Author(json_file)
#### 3- Each function will classify (generate dictionaries) of tweets respectively.
#### Example:
#### Sentiment_dictionary->{sentiment1: {tweet_id1: tweet text description } , sentiment2: { tweet_id2: tweet text description }, etc }
#### Author_dictionary->{Author1: {tweet_id1: tweet text description }, Author2: { tweet_id2: tweet text description } , etc }

```python
##########################################################
#         Solution to task 2 and 3
##########################################################
def to_Sentiment(json_file):
    d = {}
    data = json.load(json_file)
    for x in data:
        tweet_id = x
        for y in data[x]:
            sentiment = y
            for z in data[x][y]:
                content = data[x][y][z]
                if sentiment not in d:
                    d[sentiment] = {tweet_id: content}
                else:
                    d[sentiment].update({tweet_id: content})
    return d
                
##########################################################    
def to_Author(json_file):
    d = {}
    data = json.load(json_file)
    for x in data:
        tweet_id = x
        for y in data[x]:
            sentiment = y
            for z in data[x][y]:
                author = z
                content = data[x][y][z]
                if author not in d:
                    d[author] = {tweet_id: content}
                else:
                    d[author].update({tweet_id: content})
    return d
    
##########################################################
with open('output.json', 'r') as jsonfile:
    sort_by_sentiment = to_Sentiment(jsonfile)       # got dictionary return
    with open('sort_by_sentiment.txt', 'w') as sentimentfile:
        sentimentfile.write(str(sort_by_sentiment))
    with open('jsentiment.json','w') as jsentimentfile:
        var=json.dumps(sort_by_sentiment,indent=4)
        jsentimentfile.write(var)
        
# got print of dictionary in text & JSON file(to visualize) bcoz printing here very large file give "IOPub data rate exceeded."
#     print(f"sort by sentiment: \n {sort_by_sentiment} \n")
    
with open('output.json', 'r') as jsonfile:
    sort_by_author = to_Author(jsonfile)            # got dictionary return
    with open('sort_by_author.txt', 'w') as authorfile:
        authorfile.write(str(sort_by_author))
    with open('jauthor.json','w') as jauthorfile:
        var=json.dumps(sort_by_author,indent=4)
        jauthorfile.write(var)
        
# got print of dictionary in text & JSON file(to visualize) bcoz printing here very large file give "IOPub data rate exceeded."
#     print(f"sort by author \n {sort_by_author} \n")
    
    
```
